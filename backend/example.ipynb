```python
import requests

# 设置全局变量
BASE_URL = "http://localhost:5055"
JWT_TOKEN='' # 之后赋值
headers = {'Content-Type': 'application/json'}

# 注册
def register(username, password):
    url = BASE_URL + "/api/register"
    data = {'username': username, 'password': password}
    response = requests.post(url, json=data)
    return response.json()

# 登录
def login(username, password):
    url = BASE_URL + "/api/login"
    data = {'username': username, 'password': password}
    response = requests.post(url, json=data)
    jwt_token = response.cookies['jwt_token']
    return response.json(),jwt_token

# 获取 post 列表/检索 post
def get_posts(query='', page=1):
    url = BASE_URL + "/api/posts"
    cookies = {'jwt_token': JWT_TOKEN}
    params = {'query': query, 'page': page}
    response = requests.get(url, headers=headers, cookies=cookies, params=params)
    return response.json()

# 获取单一 post
def get_post_details(post_id):
    url = BASE_URL + f"/api/posts/{post_id}"
    cookies = {'jwt_token': JWT_TOKEN}
    response = requests.get(url, headers=headers, cookies=cookies)
    return response.json()

# 发表 post
def add_post(title, content):
    url = BASE_URL + "/posts"
    cookies = {'jwt_token': JWT_TOKEN}
    data = {'title': title, 'content': content}
    response = requests.post(url, headers=headers, cookies=cookies, json=data)
    return response.json()

# 发表评论
def add_comment(post_id, content):
    url = BASE_URL + f"/posts/{post_id}/comments"
    cookies = {'jwt_token': JWT_TOKEN}
    data = {'content': content}
    response = requests.post(url, headers=headers, cookies=cookies, json=data)
    return response.json()

# 获取用户 post
def get_user_posts():
    url = BASE_URL + "/api/user/posts"
    cookies = {'jwt_token': JWT_TOKEN}
    response = requests.get(url, headers=headers, cookies=cookies)
    return response.json()

# 获取收藏列表
def get_user_favorites():
    url = BASE_URL + "/api/user/favorates"
    cookies = {'jwt_token': JWT_TOKEN}
    response = requests.get(url, headers=headers, cookies=cookies)
    return response.json()

# 收藏/取消收藏
def favorate_post(post_id):
    url = BASE_URL + f"/api/posts/{post_id}/favorate" 
    cookies = {'jwt_token': JWT_TOKEN}
    response = requests.post(url, headers=headers, cookies=cookies)
    return response.json()

response = register('testuser', 'password')
print(response)
response, JWT_TOKEN = login('testuser', 'password')
print(response)
response = get_posts(query='test', page=1)
print(response)
response = add_post('Test Post', 'This is a test post.')
print(response)
response = get_posts(query='test', page=1)
print(response)
response = get_post_details(1)
print(response)
response = add_comment(1, 'This is a test comment.')
print(response)
response = get_user_posts()
print(response)
response = favorate_post(1)
print(response)
response = get_user_favorites()
print(response)
response = favorate_post(1)
print(response)
```
